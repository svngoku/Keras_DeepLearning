```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras import backend as K
```

```python
from keras.preprocessing import image
import numpy as np
```

```python
## Init VGG16 model for imagenet
model = VGG16(weights='imagenet')
```

```python
img_path = './pictures/'
```

```python
def predict_img(img_file):
    
    img = image.load_img(img_path + img_file, target_size=(224, 224, 3))
    
    x = image.img_to_array(img)
    
    x = np.expand_dims(x, axis=0)
    
    x = preprocess_input(x)
    
    preds = model.predict(x)
    
    plt.imshow(img)
    
    for prediction in decode_predictions(preds):
        print(prediction[:1])

    return preds
    
    #print(f"Prediction {decode_predictions(preds, top=3)[0]}")
```

```python
gorilla = predict_img('gorilla.jpg')
```
